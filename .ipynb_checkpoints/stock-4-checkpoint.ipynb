```python
# 마이너스값 양수로 만들어주기, 수익률은 -1보다 크기 때문에 그냥 1을 더해 양수로 만드어준다.
# (수익률 문제라 같은 값으로 더해주면 크게 상관없음)
data['re_return'] = data['return'].map(lambda x: x+1)
```

```python
#오류값 있는지 확인(inf / NaN)
if np.isinf(data['re_return']).any():
    print('inf값이 있음.')

if np.isnan(data['re_return']).any():
    print('NaN값이 있음.')

#없음
```
